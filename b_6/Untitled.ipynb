```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward(x):
    return x * w + b

def loss(x, y):
    return (y * np.log(h)) + ((1 - y) * np.log(1 - h))

def gradient(x, y):
    return (h - y) * x
```

```python
x = [2, 4, 6]
y = [0, 0, 1]
w = 2
alpha = 0.01
b = 0
loss_list = []
for epoch in range(100):
    l = 0
    for x_val, y_val in zip(x, y):
        h = sigmoid(x_val * w)
        grad = gradient(x_val, y_val)
        w = w - alpha * grad
        l += loss(x_val, y_val)
    loss_list.append(-l / len(x))
print(loss_list)
```
